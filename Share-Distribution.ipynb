```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'png' 
import sklearn 
from sklearn.manifold import TSNE 
import seaborn as sns 
from pylab import rcParams 
from sklearn import preprocessing
rcParams['figure.figsize'] = 12, 8
```

```python
df = pd.read_csv('OnlineNewsPopularity.csv').sample(n=1000)
# df = df.reindex(np.random.permutation(df.index)) # shuffling the data
df.describe()

features = df.columns[df.dtypes == np.float64]
target = ' shares'
X = preprocessing.MinMaxScaler().fit_transform(df[features])
Y = np.log10(df[target])
```

```python
from sklearn.cluster import KMeans
model = KMeans()
ll = model.fit_predict(X)
```

```python
df['cluster'] = ll
```
